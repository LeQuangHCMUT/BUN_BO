```python
data_path = '/content/drive/My Drive/BUN_BO/Data.SHL/Raw_data/'
data_name_files = [f.name for f in os.scandir(data_path) if f.is_file()]
for i in data_name_files:
  print(i)
```

```python
data_name_files = natsorted(data_name_files)  # Sắp xếp theo thứ tự số học
data_name_file_paths = []

for i in data_name_files:
    data_name_file_paths.append(os.path.join(data_path, i))

print(data_name_file_paths)
```

```python
for i in range(0, 13):
    data = pd.read_csv(data_name_file_paths[i])
    data.replace(0, pd.NA, inplace=True)
    print(data_name_file_paths[i])
    print(data.describe())
    print(data.info())
    print("#######################################################################################")
# 2,3
```
