```python
train = pd.read_csv('/content/drive/My Drive/BUN_BO/Data.SHL/Data_using/data_train.csv')
test = pd.read_csv('/content/drive/My Drive/BUN_BO/Data.SHL/Data_using/data_test.csv')

```

```python
cols_to_drop = [
    "atr01/acc_x", "atr01/acc_y", "atr01/acc_z",
    "atr01/gyro_x", "atr01/gyro_y", "atr01/gyro_z",
    "atr01/quat_w", "atr01/quat_x", "atr01/quat_y", "atr01/quat_z",
    "atr02/acc_x", "atr02/acc_y", "atr02/acc_z",
    "atr02/gyro_x", "atr02/gyro_y", "atr02/gyro_z",
    "atr02/quat_w", "atr02/quat_x", "atr02/quat_y", "atr02/quat_z",
    "atr03/acc_x", "atr03/acc_y", "atr03/acc_z",
    "atr03/gyro_x", "atr03/gyro_y", "atr03/gyro_z",
    "atr03/quat_w", "atr03/quat_x", "atr03/quat_y", "atr03/quat_z",
    "atr04/acc_x", "atr04/acc_y", "atr04/acc_z",
    "atr04/gyro_x", "atr04/gyro_y", "atr04/gyro_z",
    "atr04/quat_w", "atr04/quat_x", "atr04/quat_y", "atr04/quat_z"
]

train = train.drop(columns=cols_to_drop)
test = test.drop(columns=cols_to_drop)
```

```python
def add_difference_columns(df):
    df["Difference_atr01_atr02"] = df["atr01/acc"] - df["atr02/acc"]
    df["Difference_atr03_atr04"] = df["atr03/acc"] - df["atr04/acc"]
    df["Difference_atr01_atr03"] = df["atr01/acc"] - df["atr03/acc"]
    df["Difference_atr02_atr04"] = df["atr02/acc"] - df["atr04/acc"]
    return df

train = add_difference_columns(train)
test = add_difference_columns(test)


columns_to_drop = [col for col in train.columns if "corr" in col]
train = train.drop(columns=columns_to_drop)
test = test.drop(columns=columns_to_drop)
```
